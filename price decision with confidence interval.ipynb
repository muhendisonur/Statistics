```python
import numpy as np
import statsmodels.stats.api as sm
import random as rnd
import matplotlib.pyplot as plt
```

```python
prices = np.random.randint(15,100,1000)
```

```python
plt.figure()
plt.hist(prices) # histogram of prices
```

```python
sm.DescrStatsW(prices).tconfint_mean() # with 95% confidence rate, confidence interval applied on prices
```
